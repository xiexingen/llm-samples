1. 通过 langchain 访问 DeepSeek

```python
# 提示词模板
from langchain_core.prompts import ChatMessagePromptTemplate
# 通过 langchain_deepseek 包装了 deepseek 的访问细节
from langchain_deepseek import ChatDeepSeek
from rich import print

# 1. 创建模型的实例，langchain_deepseek 组件包里面定义好的类
llm = ChatDeepSeek(model="deepseek-chat")
# 2. 创建提示语的模板角色
prompt = "我可以如何帮助您解决{subject}的问题?"

# 3. 给提示语的模板一个角色
chat_message_prompt = ChatMessagePromptTemplate.from_template(template= prompt,role="销售顾问")

# 4. 填写模板的内容，输入一个参数 subject 解决的具体问题
# 最终的提示词是结合提示词模板加上具体的提示信息得到的
formatted_prompt = chat_message_prompt.format(subject="产品咨询")
prompts = [str(formatted_prompt)]
# 5. 让大模型进行回答
response = llm.generate(prompts)
print(response)
```
