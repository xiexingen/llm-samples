```python
# é€šè¿‡ langchain çš„ core åº“å¼•å…¥åŸºç¡€çš„æç¤ºè¯æ¨¡æ¿
from langchain_core.prompts import PromptTemplate
# å¼•å…¥ deepseek çš„ sdk
from langchain_deepseek import ChatDeepSeek
from rich import print

# åˆå§‹åŒ– deepseek çš„ sdk åˆ›å»º deepseek æ¨¡å‹çš„å®ä¾‹
llm = ChatDeepSeek(model="deepseek-chat")

# åˆ›å»ºä¸€ä¸ªé—®é¢˜æ¨¡æ¿ï¼Œå…¶ä¸­ {Query} æ˜¯å ä½ç¬¦,è¡¨ç¤ºå®¢æˆ·çš„æŸ¥è¯¢
# ç³»ç»Ÿæç¤ºè¯
# ç”¨æˆ·æç¤ºè¯ å ä½ç¬¦ {Query}
template = """
ä½ ä½œä¸ºä¸€åç»éªŒä¸°å¯Œçš„å®¢æœä»£è¡¨ï¼Œè¯·ä¸ºä»¥ä¸‹å®¢æœé—®é¢˜æä¾›è§£ç­”:
{Query}
"""

prompt = PromptTemplate(
  template=template, # å®šä¹‰é—®é¢˜æ¨¡æ¿
  input_variables=["Query"] # å®šä¹‰å ä½ç¬¦
)

# è¿™é‡Œæ˜¯å®é™…çš„ç”¨æˆ·è¾“å…¥ï¼Œä¾‹å¦‚: "å¦‚ä½•å®‰è£…"
# ç”¨æˆ·æç¤ºè¯+ç³»ç»Ÿæç¤ºè¯ =æœ€ç»ˆæç¤ºè¯
final_prompt = prompt.format(Query="å¦‚ä½•å®‰è£…")

print(f"ç»„åˆåçš„ç”¨æˆ·è¯·æ±‚:{final_prompt}")

# è°ƒç”¨ deepseek æ¨¡å‹è¿›è¡Œå›ç­”
response = llm(final_prompt)
print(response)
```

3. DeepSeek åˆ©ç”¨éƒ¨åˆ†æç¤ºè¯æ¨¡æ¿

```python
from langchain_core.prompts import PromptTemplate
from langchain_deepseek import ChatDeepSeek

llm = ChatDeepSeek(model ="deepseek-chat")

# åˆ›å»ºä¸€ä¸ª PromptTemplate å¯¹è±¡ï¼Œå®šä¹‰æ¨¡æ¿å’Œè¾“å…¥å˜é‡
# è¿™é‡Œçš„æ¨¡æ¿æ˜¯ä¸ºäº†æ„å»ºä¸€ä¸ªç”¨æˆ·è¯·æ±‚ï¼Œå…¶ä¸­{product}å’Œ{problem}æ˜¯å¾…å¡«å……çš„å ä½ç¬¦
prompt = PromptTemplate(template="æˆ‘æƒ³çŸ¥é“å…³äº{product}çš„{problem}?", input_variables = ["product","problem"])

# ä½¿ç”¨partialæ–¹æ³•ï¼Œé¢„å…ˆå¡«å……"product"å˜é‡ä¸º"ç”µè„‘"
# è¿™æ ·å¯ä»¥ä¸ºç‰¹å®šäº§å“åˆ›å»ºä¸“ç”¨çš„æç¤ºæ¨¡æ¿
partial_prompt = prompt.partial(product="ç”µè„‘")

# ä½¿ç”¨formatæ–¹æ³•ï¼Œå¡«å……å‰©ä½™çš„"problem"å˜é‡ä¸º"æ¢å¤å‡ºå‚è®¾ç½®"ï¼Œå¹¶å°†ç»“æœè½¬åŒ–ä¸ºå­—ç¬¦ä¸²
# è¿™æ ·å°±æ„å»ºäº†ä¸€ä¸ªå®Œæ•´çš„ç”¨æˆ·è¯·æ±‚ï¼šâ€œæˆ‘æƒ³çŸ¥é“å…³äºç”µè„‘çš„æ¢å¤å‡ºå‚è®¾ç½®ï¼Ÿâ€
formatted_prompt = str(partial_prompt.format(problem="æ¢å¤å‡ºå‚è®¾ç½®?"))

# å°†æ ¼å¼åŒ–çš„æç¤ºä¼ é€’ç»™LLMæ¨¡å‹å¹¶è·å–å›åº”
# è¿™é‡Œé€šè¿‡å°†æ ¼å¼åŒ–çš„æç¤ºæ”¾åœ¨åˆ—è¡¨ä¸­ä¼ é€’ç»™generateæ–¹æ³•ï¼Œä»¥è·å–æ¨¡å‹çš„å›åº”
response = llm([formatted_prompt])

print(response)
```

4. è‡ªåŠ¨å®¢æœç»„åˆ

```python
from langchain_core.prompts.pipeline import PipelinePromptTemplate
from langchain_core.prompts.prompt import PromptTemplate
from langchain_deepseek import ChatDeepSeek

llm = ChatDeepSeek(model="deepseek-chat")

# å®šä¹‰æœ€ç»ˆçš„æç¤ºæ¨¡æ¿

full_template ="""
{category_select}
{brand_select} {os_select}
{problem_select}
"""
full_prompt = PromptTemplate.from_template(full_template)


# å®šä¹‰å„ä¸ªéƒ¨åˆ†çš„æç¤ºæ¨¡æ¿
person_template = """æ‚¨å¥½ï¼Œæˆ‘æ˜¯{person}ï¼Œ"""
person_prompt = PromptTemplate.from_template(person_template)

brand_template = """æˆ‘äº†è§£åˆ°æ‚¨ä½¿ç”¨çš„æ˜¯{brand}æ‰‹æœºï¼Œ"""
brand_prompt = PromptTemplate.from_template(brand_template)

os_template = """æ“ä½œç³»ç»Ÿä¸º{os}ã€‚"""
os_prompt = PromptTemplate.from_template(os_template)

problem_template = """æ‚¨é‡åˆ°äº†{problem}çš„é—®é¢˜ï¼Œæˆ‘å¯ä»¥ä¸ºæ‚¨è§£å†³ã€‚"""
problem_prompt = PromptTemplate.from_template(problem_template)

# å°†æ‰€æœ‰éƒ¨åˆ†çš„æç¤ºæ¨¡æ¿ç»„åˆåˆ°ä¸€èµ·
input_prompts = [
  ("category_select", person_prompt),
  ("brand_select", brand_prompt),
  ("os_select", os_prompt),
  ("problem_select", problem_prompt),
]

pipeline_prompt = PipelinePromptTemplate(final_prompt= full_prompt,pipeline_prompts= input_prompts)

# æ ¼å¼åŒ–çŒåˆ°å¹¶ç”Ÿæˆ LLM å›åº”

formatted_prompt = pipeline_prompt.format(
  person="æŠ€æœ¯æ”¯æŒ",
  brand="å°ç±³",
  os="æ¾æ¹ƒç³»ç»Ÿ",
  problem="æ¢å¤å‡ºå‚è®¾ç½®"
)
response = llm.generate([str(formatted_prompt)])

print(response)
```

```python
from langchain_core.prompts import PromptTemplate,FewShotPromptTemplate
from langchain_core.example_selectors import LengthBasedExampleSelector
from langchain_deepseek import ChatDeepSeek

llm = ChatDeepSeek(model="deepseek-chat")

# åˆ›å»ºå®¢æœç›¸å…³ä¾‹å­

examples = [
  {
    "input": "å¦‚ä½•é‡ç½®å¯†ç ï¼Ÿ",
    "output": "è¯·å‰å¾€ç™»å½•é¡µé¢ï¼Œç‚¹å‡»â€œå¿˜è®°å¯†ç â€é“¾æ¥ï¼Œç„¶åæŒ‰ç…§æç¤ºæ“ä½œã€‚è¿™æ ·æ‚¨å¯ä»¥æ”¶åˆ°ä¸€ä¸ªé‡ç½®å¯†ç çš„é“¾æ¥ï¼Œé€šè¿‡é“¾æ¥è®¾ç½®æ–°çš„å¯†ç ã€‚æ˜¯å¦è§£å†³äº†æ‚¨çš„é—®é¢˜ï¼Ÿ"
  },
  {
    "input": "æ€æ ·æŸ¥è¯¢è®¢å•çŠ¶æ€ï¼Ÿ",
    "output": "è¯·ç™»å½•æ‚¨çš„è´¦æˆ·ï¼Œç„¶åç‚¹å‡»â€œæˆ‘çš„è®¢å•â€æŸ¥çœ‹è®¢å•çŠ¶æ€ã€‚åœ¨â€œæˆ‘çš„è®¢å•â€é¡µé¢ï¼Œæ‚¨å¯ä»¥çœ‹åˆ°è®¢å•çš„å½“å‰çŠ¶æ€ï¼Œä¾‹å¦‚â€œå¤„ç†ä¸­â€ï¼Œâ€œå·²å‘è´§â€æˆ–â€œå·²å®Œæˆâ€ã€‚è¿™å¯¹æ‚¨æœ‰å¸®åŠ©å—ï¼Ÿ"
  },
  {
    "input": "å¦‚ä½•ç”³è¯·é€€è´§ï¼Ÿ",
    "output": "è¯·ç™»å½•æ‚¨çš„è´¦æˆ·ï¼Œæ‰¾åˆ°ç›¸åº”çš„è®¢å•ï¼Œç‚¹å‡»â€œç”³è¯·é€€è´§â€æŒ‰é’®ã€‚ç³»ç»Ÿå°†å¼•å¯¼æ‚¨å®Œæˆé€€è´§ç”³è¯·æµç¨‹ï¼ŒåŒ…æ‹¬æä¾›é€€è´§ç†ç”±å’Œé€€è´§åœ°å€ã€‚éœ€è¦æˆ‘ä¸ºæ‚¨æä¾›å…¶ä»–å¸®åŠ©å—ï¼Ÿ"
  },
  {
    "input": "å¦‚ä½•ä½¿ç”¨ä¼˜æƒ åˆ¸ï¼Ÿ",
    "output": "åœ¨ç»“è´¦é¡µé¢ï¼Œæ‰¾åˆ°â€œä½¿ç”¨ä¼˜æƒ åˆ¸â€é€‰é¡¹ï¼Œè¾“å…¥ä¼˜æƒ åˆ¸ä»£ç ã€‚ä¼˜æƒ åˆ¸å°†ä¼šç«‹å³ç”Ÿæ•ˆï¼Œæ‚¨å¯ä»¥çœ‹åˆ°æ€»ä»·å·²ç»è°ƒæ•´ã€‚è¿˜æœ‰å…¶ä»–æˆ‘å¯ä»¥ä¸ºæ‚¨åšçš„å—ï¼Ÿ"
  },
]

example_prompt = PromptTemplate(
  input_variables = ["input","output"],
  template = "é—®é¢˜:{input}\nå›ç­”:{output}\n"
)

# ç”Ÿæˆä¾‹å­é€‰æ‹©çš„å®ä¾‹ï¼Œä¾‹å­å’Œä¾‹å­æ¨¡ç‰ˆ
example_selector = LengthBasedExampleSelector(
  examples = examples,
  example_prompt = example_prompt
)

dynamic_prompt = FewShotPromptTemplate(
  #é€šè¿‡example_selectorçš„ä¾‹å­ï¼Œå”¤é†’deepseekåœ¨å®¢æˆ·æœåŠ¡è¯æœ¯æ–¹é¢çš„è®°å¿†
  example_selector=example_selector,
  example_prompt=example_prompt,
  #ç³»ç»Ÿæç¤ºè¯
  prefix="å›ç­”ä»¥ä¸‹å®¢æˆ·çš„é—®é¢˜ï¼Œå¹¶æä¾›é¢å¤–çš„è§£é‡Šæˆ–ä¿¡æ¯ï¼Œç„¶åè¯¢é—®ä»–ä»¬æ˜¯å¦æ»¡æ„ã€‚",
  suffix="é—®é¢˜: {query}\nå›ç­”:",
  input_variables=["query"],
)

# ç¤ºä¾‹è¾“å…¥
print(dynamic_prompt.format(query="å¦‚ä½•å–æ¶ˆè®¢å•ï¼Ÿ"))

# ä½ å¯ä»¥ä½¿ç”¨ä½ çš„è¯­è¨€æ¨¡å‹æ¥ç”Ÿæˆè¾“å‡º
try:
    output = llm.generate([dynamic_prompt.format(query="å¦‚ä½•å–æ¶ˆè®¢å•ï¼Ÿ")])
    print(output)
except Exception as e:
    import traceback
    traceback.print_exc()
```
